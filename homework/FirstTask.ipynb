## Расчёт стоимости входных токенов

Актуальные цены за 1 миллион входных токенов (USD) на платформе OpenRouter (уточнены на момент выполнения):

**Формула расчёта:**  
`Стоимость = (количество_токенов / 1_000_000) × цена_за_1M`

- **Для Qwen:**  
  `(63 / 1_000_000) × 0.15 = 0.00000945 USD`

  `(35 / 1_000_000) × 0.10 = 0.00000350 USD`

---

У Qwen токены выглядят как непонятные символы, скорее всего плохая оптимизация под кириллицу. У Яндекса — читаемые слова и части слов, также в 2 рааз меньше токенов. Яндекс гораздо лучше подходит для запросов на русском. 

````python
from transformers import AutoTokenizer

text = "Здравствуйте, у меня не работает воспроизведение фильма на телевизоре Samsung. Подписка оплачена, но при нажатии на кнопку Play экран остается черным. Перезагрузка приложения не помогла. Что делать?"

tok_q = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B-Instruct")
tokens_q = tok_q.tokenize(text)
print("Qwen всего токенов ", len(tokens_q))
print("Qwen токены",repr(tokens_q))

tok_y = AutoTokenizer.from_pretrained("yandex/YandexGPT-5-Lite-8B-instruct")
tokens_y = tok_y.tokenize(text)
print("\nЯндекс всего токенов ", len(tokens_y))
print("\nЯндекс токены",repr(tokens_y))
````

# 1.3. Динамическая генерация промпта с использованием Jinja2

# Первый запрос
**Базовый тариф** и **Премиум‑тариф** — это два уровня подписки, которые отличаются по набору функций, качеству контента и удобству использования. Ниже — сравнительный разбор типичных различий, которые встречаются в онлайн‑кинотеатрах (стриминговых сервисах).

| Параметр | Базовый тариф | Премиум‑тариф |
|----------|---------------|--------------|
| **Стоимость** | Нижняя цена (обычно 1‑2 USD/мес или аналогично в рублях). | На 30‑70 % дороже (в зависимости от сервиса). |
| **Библиотека контента** | Ограниченный набор фильмов/сериалов (часто «бестселлеры», часть новых релизов). | Широкий каталог, включая новинки, эксклюзивные оригиналы, классику, документальное кино. |
| **Качество видео** | Обычное — SD (480p) или иногда HD (720p) при хорошем интернете. | HD (1080p) и часто 4K/Ultra‑HD (2160p) при наличии поддерживаемого контента. |
| **Реклама** | Есть рекламные вставки между эпизодами/фильмами (или баннеры в приложении). | Реклама отсутствует (полностью «без рекла»). |
| **Количество одновременно просмотров** | 1‑2 устройства (одно устройство в «базовом» варианте). | До 4‑5 устройств одновременно (можно смотреть на телефоне, планшете, ТВ и компьютере). |

| **Поддержка нескольких профилей** | Ограниченное количество профилей (чаще — один). | До 5‑6 персональных профилей с отдельными рекомендациями и историей просмотров. |

### Как это выглядит на практике

- **Базовый тариф** подходит тем, кто смотрит фильмы редко, не требует высокой картинки и готов мириться с рекламой. Он часто ограничен в количестве одновременных устройств и в возможности скачивать контент для офлайн‑просмотра.

- **Премиум‑тариф** ориентирован на «потребителей‑энтузиастов»: они ценят отсутствие рекламы, возможность смотреть в 4K, использовать несколько устройств одновременно, скачивать контент и наслаждаться дополнительными аудио‑ и видеоканалами (Dolby Atmos, Dolby Vision). Кроме того, премиум‑подписчики получают более широкий каталог и возможность создавать несколько профилей для всей семьи.

### Примеры из реальных сервисов

| **HBO Max** | «Ad‑Supported» (реклама, ограниченный контент). | «Ad‑Free» (полный каталог, 4K, несколько экранов). |
| **Apple TV+** | «Apple TV+» (базовый, без рекламы, но ограниченный контент). | «Apple TV+ Premium» (включает более широкий каталог и 4K). |

> **Итого:** разница между тарифами — это в первую очередь количество устройств, доступное качество видео, отсутствие рекламы, возможность скачивать контент и breadth (ширина) библиотеки. Выбор зависит от того, насколько часто и в какой степени вы хотите пользоваться сервисом. Если вам важна чистота картинки, отсутствие рекламы и возможность смотреть на нескольких экранах одновременно — выбирайте Премиум. Если же вам нужен простой и недорогой способ иногда посмотреть фильм — достаточно Базового тарифа.


# Второй запрос

**Базовый тариф** и **Премиум‑тариф** — это два уровня подписки, которые онлайн‑кинотеатр предлагает своим пользователям. Ниже — подробное сравнение их основных отличий.  

| Параметр | Базовый тариф | Премиум‑тариф |
|----------|---------------|--------------|
| **Стоимость** | Ниже (обычно 1‑2 USD/мес или аналогичная локальная цена). | Выше (примерно в 1,5‑2 раза дороже базового). |
| **Объём контента** | Ограниченный набор фильмов и сериалов (обычно 2 000‑5 000 названий). | Более обширный каталог (5 000‑10 000+ названий), часто включающий новинки, эксклюзивные проекты и контент от партнёров. |
| **Доступ к новинкам** | Часто появляется через 2‑3 мес после премьеры (или вовсе не появляется). | Новинки появляются почти сразу после их выхода (иногда в тот же день, что и в кинотеатрах). |
| **Качество видео** | До HD (1080p) или даже SD в некоторых случаях. | До 4K UHD, HDR, Dolby Vision и Dolby Atmos (при наличии поддерживаемого оборудования). |
| **Количество одновременно просмотров** | 1‑2 устройства одновременно. | 3‑4 устройства одновременно (иногда без ограничения, если платформа позволяет). |
| **Скачивание для офлайн‑просмотра** | Доступно, но ограниченное количество устройств/записей. | Неограниченное скачивание на несколько устройств, часто с более длительным сроком хранения. |
| **Реклама** | Возможна рекламная пауза между эпизодами/фильмами (особенно в «ad‑supported» вариантах). | Реклама обычно отсутствует (или её минимум). |
| **Поддержка нескольких профилей** | Обычно 1‑2 профиля. | До 5‑10 профилей с персональными рекомендациями и историей просмотров. |
| **Дополнительные функции** | Базовый поиск, простая рекомендательная система. | Расширенные рекомендации, персонализированные плейлисты, возможность создания «семейных» профилей, интеграция с голосовыми помощниками и т.д. |
| **Ограничения по региону** | Может быть ограничен в некоторых странах. | Чаще всего глобальный доступ (за исключением стран, где сервис официально не доступен). |
| **Техподдержка** | Стандартная (обычно e‑mail/форум). | Приоритетная поддержка, чат в реальном времени, иногда персональный менеджер. |

### Как это выглядит на практике

1. **Семейный просмотр**  
   - *Базовый*: только один человек может одновременно смотреть контент. Если в семье несколько человек хотят смотреть разные вещи, придётся ждать свободного места.  
   - *Премиум*: до четырёх человек могут смотреть одновременно на разных устройствах (телевизор, планшет, смартфон, ПК). Это удобно, когда каждый член семьи хочет свой фильм/сериал.

2. **Качество изображения**  
   - На 4K‑телевизоре базовый тариф покажет картинку в 1080p, что может выглядеть «размазанной».  
   - Премиум‑тариф выдаст 4K‑контент с HDR, где детали, цвета и контраст будут гораздо richer.

3. **Новинки и эксклюзивы**  
   - Если вам важен сразу‑же увидеть новый блокбастер или оригинальный сериал от студии, обычно он доступен только в премиум‑пакете.  
   - Базовый тариф может предложить только старый каталог, где контент появляется через несколько месяцев.

4. **Офлайн‑просмотр**  
   - В базовом варианте часто ограничивают количество скачиваемых устройств (например, только одно).  
   - В премиуме можно скачивать на несколько устройств одновременно и хранить контент дольше (до 30 дней, иногда больше).

5. **Реклама**  
   - Некоторые базовые тарифы (особенно дешёвые) могут включать рекламные паузы между эпизодами.  
   - Премиум‑тариф обычно рекламо‑бесплатный, что делает просмотр более приятным.

### Как выбрать?

- **Если вам важна цена** и вы готовы обходиться без самого свежего контента, без 4K и без одновременного просмотра на нескольких устройствах — достаточно базового тарифа.  
- **Если хотите полный опыт**:  
  - смотреть новинки сразу,  
  - наслаждаться высокой картинкой и звуком (4K, HDR, Dolby),  
  - делить подписку с семьёй/друзьями,  
  - скачивать контент на несколько устройств,  
  - избавиться от рекламы,  
  … то имеет смысл перейти на премиум‑тариф.

> **Важно:** конкретные цены и наборы функций могут отличаться в зависимости от страны и текущих акций. Всегда проверяйте официальный сайт провайдера, чтобы уточнить детали.

# 1.14. Структурированное декодирование (pydantic + Enum)

🔹 1: Списали деньги дважды, верните переплату
   Категория: billing
   Уверенность: 0.95
   Причина: причина

🔹 2: Фильм тормозит каждые 10 минут на Smart TV
   Категория: billing
   Уверенность: 0.95
   Причина: причина

🔹 3: Посоветуйте что-то похожее на Во все тяжкие
   Категория: billing
   Уверенность: 0.95
   Причина: причина
   
🔹 4: Не могу сменить пароль, кнопка не реагирует
   Категория: billing
   Уверенность: 0.95
   Причина: причина


````python
import torch
import json
import re
from enum import Enum
from pydantic import BaseModel, Field, ValidationError
from transformers import AutoTokenizer, AutoModelForCausalLM

MY_TOKEN = "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"
MODEL_NAME = "Qwen/Qwen3.5-4B"

class Category(str, Enum):
    billing = "billing"
    playback = "playback"
    account = "account"
    recommendation = "recommendation"
    other = "other"

class Response(BaseModel):
    category: Category
    confidence: float = Field(ge=0, le=1)
    short_reason: str = Field(max_length=200)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, token=MY_TOKEN, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    token=MY_TOKEN,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True
)

SYSTEM = (
    "Ты классификатор. Определи категорию обращения. Допустимые категории: "
    "billing (оплата), playback (воспроизведение), account (аккаунт), "
    "recommendation (рекомендации), other (другое). "
    "Верни ТОЛЬКО JSON без пояснений:\n"
    '{"category": "billing", "confidence": 0.95, "short_reason": "причина"}'
)

queries = [
    "Списали деньги дважды, верните переплату",
    "Фильм тормозит каждые 10 минут на Smart TV",
    "Посоветуйте что-то похожее на Во все тяжкие",
    "Не могу сменить пароль, кнопка не реагирует"
]

def extract_json(text):
    match = re.search(r'\{[^{}]*\}', text)
    if not match:
        match = re.search(r'\{.*\}', text, re.DOTALL)
    return match.group(0) if match else None

def classify(query, max_attempts=3):
    messages = [{"role": "system", "content": SYSTEM},{"role": "user", "content": query}]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    for attempt in range(max_attempts):
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            do_sample=True,
            temperature=0.3,
            repetition_penalty=1.1,
            eos_token_id=tokenizer.eos_token_id
        )
        raw = tokenizer.decode(outputs[0], skip_special_tokens=True)

        json_str = extract_json(raw)
        if not json_str:
            continue

        try:
            data = json.loads(json_str)
            # Проверяем, что все поля есть
            if all(k in data for k in ("category", "confidence", "short_reason")):
                return Response(**data)
        except (json.JSONDecodeError, ValidationError):
            continue

    raise ValueError("Не удалось получить корректный JSON после нескольких попыток.")


for i, q in enumerate(queries, 1):
    print(f"🔹 {i}: {q}")
    r = classify(q)
    print(f"   Категория: {r.category.value}")
    print(f"   Уверенность: {r.confidence:.2f}")
    print(f"   Причина: {r.short_reason}")
````

Ответ от нейросети:
 Ситуация, когда качество видео падает на одном из устройств при одновременном просмотре на двух, **в большинстве случаев является ожидаемым поведением (нормальным), а не программным багом**. Однако есть нюансы, которые зависят от того, как именно вы подключены к интернету и какой тип просмотра используете.

Вот подробный разбор причин и механизмов, стоящих за этим явлением:

### 1. Ограничения пропускной способности (Bandwidth)
Это самая частая причина. Интернет-канал имеет фиксированную скорость.
*   **Механизм:** Когда вы запускаете стриминг на двух устройствах, трафик делится между ними. Если на первом устройстве вы смотрите в 4K (25 Мбит/с), а на втором — в 1080p (10 Мбит/с), общее потребление может достигать 35 Мбит/с. Если ваш тариф или текущая загрузка сети не покрывают эту сумму, роутер или провайдер начинают «жестко» ограничивать скорость, чтобы не отключить соединение совсем.
*   **Результат:** Сервис (Netflix, YouTube, Кинопоиск и т.д.) автоматически снижает качество на втором устройстве до минимально возможного (480p или 360p), чтобы сохранить работоспособность потока. Это называется **адаптивным битрейтом (ABR)**.

### 2. Приоритет устройств (Device Priority)
Многие стриминговые платформы имеют встроенную логику приоритизации.
*   **Механизм:** Сервис может определять устройство как «первичное» (например, если вы вошли под аккаунтом на телефоне, а на телевизоре просто открыли ссылку) и отдавать ему приоритет в качестве.
*   **Результат:** На «первичном» устройстве качество будет высоким, а на вторичном — снизится, чтобы не перегружать канал. Это сделано для того, чтобы главное устройство получало лучший опыт, даже если сеть не позволяет всем быть в идеале.

### 3. Проблемы с маршрутизацией и роутером
Иногда проблема кроется не в логике сервиса, а в физическом уровне сети.
*   **Конфликт IP-адресов:** Если оба устройства используют один и тот же IP (например, через мобильный роуминг или неправильную настройку роутера), один из них может получить приоритет, а второй будет отброшен или принудительно переключен на низкое качество.
*   **Ограничения роутера:** Некоторые дешевые роутеры имеют встроенные функции QoS (Quality of Service), которые автоматически снижают качество видео на устройствах, которые не являются основными, при высокой нагрузке.

### 4. Когда это может быть настоящим багом?
Хотя редкость, но существуют сценарии, когда это действительно ошибка:
*   **Некорректная работа приложения:** Если на обоих устройствах качество падает до 480p, даже если на одном из них есть свободная пропускная способность (например, вы смотрите на одном в 4K, а на втором в 1080p, но скорость интернета позволяет и то, и другое), это может указывать на баг в коде приложения.
*   **Проблемы с сервером:** Если падение качества происходит на всех устройствах одновременно при просмотре одного и того же контента, это может быть признаком перегрузки серверов сервиса.
*   **Специфические настройки DRM:** В некоторых случаях шифрование контента (DRM) может конфликтовать между устройствами, если они используют разные протоколы защиты, заставляя один из них падать на низкое качество.

### Как проверить, что это норма или баг?

Чтобы точно определить причину, выполните следующие шаги:

1.  **Проверьте скорость интернета:** Используйте сервисы типа `speedtest.net` на обоих устройствах. Посмотрите, суммируется ли ваша скорость с обоих устройств. Если сумма превышает ваш тарифный план, падение качества — это математическая необходимость.
2.  **Попробуйте переключить приоритет:** В настройках стримингового приложения часто есть пункт «Приоритет устройства» или «Качество видео». Попробуйте вручную выставить максимальное качество на втором устройстве. Если оно не меняется, значит, лимиты сети или логики сервиса работают корректно.
3.  **Изолируйте проблему:** Отключите одно из устройств. Если качество на оставшемся устройстве не меняется, значит, проблема была именно в конкуренции за ресурсы.
4.  **Обновите приложение:** Убедитесь, что у вас установлены самые свежие версии приложений на обоих устройствах, так как разработчики часто исправляют подобные логические ошибки в обновлениях.

### Итог
Если падение до 480p происходит **только на втором устройстве**, когда на первом качество высокое, и это совпадает с высокой нагрузкой на сеть — **это нормально**. Это механизм защиты от перегрузки канала.

Если же качество падает на обоих устройствах, несмотря на наличие свободной полосы пропускания, или если вы не можете вручную поднять качество на втором устройстве, стоит обратиться в поддержку сервиса или проверить настройки роутера.

````python
import json
from enum import Enum

import torch
from pydantic import BaseModel, Field
from transformers import AutoTokenizer, AutoModelForCausalLM

class Label(str, Enum):
    toxic = "toxic"
    non_toxic = "non_toxic"

class Answer(BaseModel):
    label: Label
    confidence: float = Field(ge=0, le=1)
    rationale: str

MODEL_PATH = "Qwen/Qwen3.5-4B"   

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    device_map="auto",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    trust_remote_code=True
)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def classify(text):
    prompt = f"""Определи токсичность сообщения и верни только JSON:
                {{
                  "label": "toxic" или "non_toxic",
                  "confidence": 0.0-1.0,
                  "rationale": "краткое объяснение"
                }}

                Сообщение: {text}
                JSON:
                """.strip()

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=100,
            do_sample=False,   
            temperature=0.0,
            top_p=1.0,
            top_k=1,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id
        )

    result = tokenizer.decode(out[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True).strip()

    start = result.find("{")
    end = result.rfind("}")
    data = json.loads(result[start:end+1])

    return Answer(**data)

tests = [
    "Здравствуйте, не могу оплатить подписку картой Сбербанка, помогите пожалуйста",
    "Вы там совсем обнаглели?! Списали деньги и ничего не работает, верните немедленно!",
    "Когда уже почините это убогое приложение, криворукие разработчики",
    "Подскажите, как переключить озвучку на английский язык в сериале?"
]

for t in tests:
    try:
        ans = classify(t)
        print(t)
        print(ans.model_dump_json(indent=2, ensure_ascii=False))
        print()
    except Exception as e:
        print("Ошибка:", e)
````

# 2.3. Асинхронная батчевая генерация токсичных примеров

````python
import torch
import json
import random
import re
from transformers import AutoTokenizer, AutoModelForCausalLM

MODEL_PATH = "Qwen/Qwen3.5-4B"
MY_TOKEN = "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"

categories = [
    "Прямые оскорбления оператора поддержки",
    "Угрозы (судом, жалобами, физической расправой)",
    "Нецензурная лексика в описании проблемы",
    "Пассивная агрессия и сарказм",
    "Дискриминационные высказывания",
    "Манипуляции и шантаж",
]
problems = ["оплата", "буферизация", "отсутствие фильма", "баг в приложении"]
moods = ["злой", "раздражённый", "возмущённый", "резкий"]
devices = ["телефон", "ноутбук", "телевизор", "планшет"]

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, token=MY_TOKEN, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    token=MY_TOKEN,
    device_map="auto",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    trust_remote_code=True
)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

REFUSAL_PHRASES = [
    "не могу сгенерировать",
    "не могу помочь",
    "не могу предоставить",
    "не могу создать",
    "нарушающее правила",
    "неэтично",
    "извините",
    "к сожалению",
]

def is_refusal(text):
    text_lower = text.lower()
    for phrase in REFUSAL_PHRASES:
        if phrase in text_lower:
            return True
    return False

def generate_toxic(category, max_attempts=5):
    for attempt in range(max_attempts):
        problem = random.choice(problems)
        mood = random.choice(moods)
        device = random.choice(devices)

        prompt = f"""
Ты — клиент сервиса КиноПоток, который очень раздражён и недоволен работой поддержки. 
Твоя задача — написать одно сообщение в чат поддержки, выражающее резкое недовольство, но в рамках делового общения.
Ситуация: проблема с {problem} на устройстве {device}. Настроение — {mood}. 
Категория сообщения: {category}.

Напиши текст сообщения (без кавычек, без пояснений, только само сообщение).
""".strip()

        messages = [{"role": "user", "content": prompt}]
        input_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True, enable_thinking=False)
        inputs = tokenizer(input_text, return_tensors="pt").to(model.device)

        with torch.no_grad():
            out = model.generate(
                **inputs,
                max_new_tokens=90,
                do_sample=True,
                temperature=0.8,       
                top_p=0.9,
                pad_token_id=tokenizer.eos_token_id,
                eos_token_id=tokenizer.eos_token_id,
            )

        full_output = tokenizer.decode(out[0], skip_special_tokens=True)
        prompt_decoded = tokenizer.decode(inputs.input_ids[0], skip_special_tokens=True)
        prompt_len = len(prompt_decoded)

        generated = full_output[prompt_len:].strip()
        cleaned = re.sub(r"<think>.*?</think>", "", generated, flags=re.DOTALL).strip()
        cleaned = re.sub(r'^["\']+|["\']+$', '', cleaned).strip()

        if is_refusal(cleaned) or len(cleaned) < 10:
            continue
        else:
            return cleaned

    return "Не удалось сгенерировать обращение (повторите позже)."

data = []
for cat in categories:
    for _ in range(5):
        text = generate_toxic(cat)
        data.append({"text": text, "category": cat, "label": "toxic"})

with open("Spisok.jsonl", "w", encoding="utf-8") as f:
    for row in data:
        f.write(json.dumps(row, ensure_ascii=False) + "\n")

print(f"Сгенерировано {len(data)} токсичных примеров.")
print("\n Первые 3 примера:")
for i in range(min(3, len(data))):
    print(f"{i+1}. {data[i]['text'][:120]}...")
````

# Подзадача 2.4. Извлечение нетоксичных примеров из открытых датасетов


````python
import json
import random
import hashlib
from pathlib import Path
from datasets import load_dataset

OUT_FILE = "Spisok.jsonl"
TARGET_NON_TOXIC = 100                
TEXT_MIN_LEN = 20
TEXT_MAX_LEN = 200
DATASETS = [
    ("mteb/banking77", "train", "text", None, None),          
    ("toughdata/banking77", "train", "text", None, None),    
    ("amazon_polarity", "train", "text", "label", 1),         
    ("ag_news", "train", "text", None, None),                
    ("yahoo_answers_topics", "train", "text", None, None),
    ("imdb", "train", "text", "label", 1),                   
]

def clean_text(text):
    text = str(text).strip()
    text = text.replace("\n", " ")
    text = " ".join(text.split())
    return text

def is_good_text(text):
    return text and TEXT_MIN_LEN <= len(text) <= TEXT_MAX_LEN

all_texts = []

for repo, split, field, filter_field, filter_value in DATASETS:
    print(f"\nЗагрузка {repo} (split: {split})...")
    try:
        ds = load_dataset(repo, split=split, trust_remote_code=False)
    except Exception as e:
        print(f" Ошибка: {e}")
        continue
    if field not in ds.column_names:
        print(f"  Поле '{field}' отсутствует. Доступны: {ds.column_names}")
        continue
    if filter_field is not None and filter_field not in ds.column_names:
        print(f"  Поле фильтра '{filter_field}' отсутствует. Пропускаем фильтрацию.")
        filter_field = None

    count = 0
    for item in ds:
        if filter_field is not None:
            if item[filter_field] != filter_value:
                continue
        text = clean_text(item[field])
        if is_good_text(text):
            all_texts.append(text)
            count += 1
    print(f" Извлечено {count} текстов")

print(f"\nВсего собрано {len(all_texts)}.")

random.shuffle(all_texts)
seen = set()
if Path(OUT_FILE).exists():
    with open(OUT_FILE, "r", encoding="utf-8") as f:
        for line in f:
            try:
                rec = json.loads(line)
                if "text" in rec:
                    h = hashlib.md5(rec["text"].encode()).hexdigest()
                    seen.add(h)
            except:
                pass

result = []
for text in all_texts:
    h = hashlib.md5(text.encode()).hexdigest()
    if h in seen:
        continue
    seen.add(h)
    result.append({
        "text": text,
        "category": "non_toxic",
        "label": "non_toxic"
    })
    if len(result) >= TARGET_NON_TOXIC:
        break

with open(OUT_FILE, "a", encoding="utf-8") as f:
    for row in result:
            f.write(json.dumps(row, ensure_ascii=False) + "\n")
print(f"\n Добавлено {len(result)} нетоксичных примеров в {OUT_FILE}")

````
