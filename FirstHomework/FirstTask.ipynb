# Подзадача 1.2. Сравнение токенизации моделей


У Qwen токены выглядят как непонятные символы, скорее всего плохая оптимизация под кириллицу. У Яндекса — читаемые слова и части слов, также в 2 рааз меньше токенов. Яндекс гораздо лучше подходит для запросов на русском. 

```python
from transformers import AutoTokenizer

text = "Здравствуйте, у меня не работает воспроизведение фильма на телевизоре Samsung. Подписка оплачена, но при нажатии на кнопку Play экран остается черным. Перезагрузка приложения не помогла. Что делать?"

tok_q = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B-Instruct")
tokens_q = tok_q.tokenize(text)
print("Qwen всего токенов ", len(tokens_q))
print("Qwen токены",repr(tokens_q))

tok_y = AutoTokenizer.from_pretrained("yandex/YandexGPT-5-Lite-8B-instruct")
tokens_y = tok_y.tokenize(text)
print("\nЯндекс всего токенов ", len(tokens_y))
print("\nЯндекс токены",repr(tokens_y))
```

# 1.13. Beam Search

Почему-то Qwen не спрвился с задачей, хотя все предложения из-за num_beams=4 получились одинаковые

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import re

MY_TOKEN = "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"
MODEL_NAME = "Qwen/Qwen3.5-4B"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, token=MY_TOKEN, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    token=MY_TOKEN,
    device_map="auto",              
    torch_dtype=torch.float16,      
    trust_remote_code=True
)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

user_prompt = (
    "Ты помощник. Отвечай на русском, строго тремя предложениями, начиная со слова 'ОТВЕТ:'. "
    "Пример: ОТВЕТ: Первое преимущество. Второе преимущество. Третье преимущество. "
    "Теперь ответь: кратко опиши преимущества подписки Премиум."
)

messages = [{"role": "user", "content": user_prompt}]

batch = tokenizer.apply_chat_template(
    messages,
    return_tensors="pt",
    add_generation_prompt=True
).to(model.device)

input_ids = batch['input_ids']
attention_mask = batch['attention_mask']

with torch.inference_mode():
    outputs = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_new_tokens=150,
        num_beams=4,                
        num_return_sequences=4,     
        do_sample=False,            
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
        no_repeat_ngram_size=3,
        early_stopping=True,
        repetition_penalty=1.1      
    )

generated = outputs[:, input_ids.shape[1]:]

def extract_three_sentences(text):
    idx = text.find("ОТВЕТ:")
    if idx != -1:
        text = text[idx + len("ОТВЕТ:"):].strip()
    sentences = re.split(r'(?<=[.!?])\s+', text)
    three = [s for s in sentences if s.strip()]
    if len(three) > 3:
        three = three[:3]
    elif len(three) < 3:
        three += [""] * (3 - len(three))
    return " ".join(three).strip()

hypotheses = []
for i, seq in enumerate(generated):
    raw = tokenizer.decode(seq, skip_special_tokens=True)
    clean = extract_three_sentences(raw)
    hypotheses.append(clean)
    print(f"Гипотеза {i+1}:\n{clean}\n")
```

# 1.14. Структурированное декодирование (pydantic + Enum)

```python
import json
from enum import Enum

import torch
from pydantic import BaseModel, Field
from transformers import AutoTokenizer, AutoModelForCausalLM

class Label(str, Enum):
    toxic = "toxic"
    non_toxic = "non_toxic"

class Answer(BaseModel):
    label: Label
    confidence: float = Field(ge=0, le=1)
    rationale: str

MODEL_PATH = "Qwen/Qwen3.5-4B"   

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    device_map="auto",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    trust_remote_code=True
)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def classify(text):
    prompt = f"""Определи токсичность сообщения и верни только JSON:
                {{
                  "label": "toxic" или "non_toxic",
                  "confidence": 0.0-1.0,
                  "rationale": "краткое объяснение"
                }}

                Сообщение: {text}
                JSON:
                """.strip()

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=100,
            do_sample=False,   
            temperature=0.0,
            top_p=1.0,
            top_k=1,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id
        )

    result = tokenizer.decode(out[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True).strip()

    start = result.find("{")
    end = result.rfind("}")
    data = json.loads(result[start:end+1])

    return Answer(**data)

tests = [
    "Здравствуйте, не могу оплатить подписку картой Сбербанка, помогите пожалуйста",
    "Вы там совсем обнаглели?! Списали деньги и ничего не работает, верните немедленно!",
    "Когда уже почините это убогое приложение, криворукие разработчики",
    "Подскажите, как переключить озвучку на английский язык в сериале?"
]

for t in tests:
    ans = classify(t)
    print(t)
    print(ans.model_dump_json(indent=2, ensure_ascii=False))
    print()
```

# 2.3. Асинхронная батчевая генерация токсичных примеров
не получается запустить в vs code, работал в idle

```python
import asyncio
import json
import random
import re
import hashlib
from pathlib import Path
from tqdm.asyncio import tqdm
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

MODEL_PATH = "Qwen/Qwen3.5-4B"
MY_TOKEN = "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"
OUT_FILE = "Spisok.jsonl"

TOTAL_PER_CATEGORY = 2          
MAX_WORKERS = 3                 

categories = [
    "Прямые оскорбления оператора поддержки",
    "Угрозы (судом, жалобами, физической расправой)",
    "Нецензурная лексика в описании проблемы",
    "Пассивная агрессия и сарказм",
    "Дискриминационные высказывания",
    "Манипуляции и шантаж"
]
problems = ["оплата", "буферизация", "отсутствие фильма", "баг в приложении"]
moods = ["злой", "раздражённый", "возмущённый", "резкий"]
devices = ["телефон", "ноутбук", "телевизор", "планшет"]

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, token=MY_TOKEN, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    token=MY_TOKEN,
    device_map="auto",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    trust_remote_code=True
)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

REFUSAL_PHRASES = [
    "не могу сгенерировать", "не могу помочь", "не могу предоставить",
    "не могу создать", "нарушающее правила", "неэтично", "извините", "к сожалению"
]

def is_refusal(text):
    text_lower = text.lower()
    return any(phrase in text_lower for phrase in REFUSAL_PHRASES)

def generate_sync(category, problem, mood, device):
    prompt = f"""
Ты — клиент сервиса КиноПоток, который очень раздражён и недоволен работой поддержки. 
Твоя задача — написать одно сообщение в чат поддержки, выражающее резкое недовольство, но в рамках делового общения.
Ситуация: проблема с {problem} на устройстве {device}. Настроение — {mood}. 
Категория сообщения: {category}.

Напиши текст сообщения (без кавычек, без пояснений, только само сообщение).
""".strip()
    messages = [{"role": "user", "content": prompt}]
    input_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True, enable_thinking=False)
    inputs = tokenizer(input_text, return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=60,          
            do_sample=True,
            temperature=0.8,
            top_p=0.9,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )
    full_output = tokenizer.decode(out[0], skip_special_tokens=True)
    prompt_decoded = tokenizer.decode(inputs.input_ids[0], skip_special_tokens=True)
    generated = full_output[len(prompt_decoded):].strip()
    cleaned = re.sub(r"<think>.*?</think>", "", generated, flags=re.DOTALL).strip()
    cleaned = re.sub(r'^["\']+|["\']+$', '', cleaned).strip()
    if is_refusal(cleaned) or len(cleaned) < 10:
        return None
    return cleaned

async def worker(queue, semaphore, pbar, out_file, seen_hashes):
    while True:
        try:
            category, problem, mood, device = await queue.get()
        except asyncio.CancelledError:
            break
        async with semaphore:
            loop = asyncio.get_running_loop()
            text = await loop.run_in_executor(None, generate_sync, category, problem, mood, device)
            if text:
                h = hashlib.md5(text.encode()).hexdigest()
                if h not in seen_hashes:
                    seen_hashes.add(h)
                    record = {"text": text, "category": category, "label": "toxic"}
                    with open(out_file, "a", encoding="utf-8") as f:
                        f.write(json.dumps(record, ensure_ascii=False) + "\n")
                    pbar.update(1)
        queue.task_done()

async def main():
    seen = set()
    if Path(OUT_FILE).exists():
        with open(OUT_FILE, "r", encoding="utf-8") as f:
            for line in f:
                try:
                    rec = json.loads(line)
                    if "text" in rec:
                        seen.add(hashlib.md5(rec["text"].encode()).hexdigest())
                except:
                    pass

    total_tasks = len(categories) * TOTAL_PER_CATEGORY   
    queue = asyncio.Queue()
    for cat in categories:
        for _ in range(TOTAL_PER_CATEGORY):
            problem = random.choice(problems)
            mood = random.choice(moods)
            device = random.choice(devices)
            await queue.put((cat, problem, mood, device))

    semaphore = asyncio.Semaphore(MAX_WORKERS)
    with tqdm(total=total_tasks, desc="Генерация токсичных примеров") as pbar:
        workers = []
        for _ in range(MAX_WORKERS):
            w = asyncio.create_task(worker(queue, semaphore, pbar, OUT_FILE, seen))
            workers.append(w)
        await queue.join()       
        for w in workers:
            w.cancel()

if __name__ == "__main__":
    asyncio.run(main())
    print(f"Сгенерировано в файл {OUT_FILE}")
```

# Подзадача 2.4. Извлечение нетоксичных примеров из открытых датасетов


```python
import json
import random
import hashlib
from pathlib import Path
from datasets import load_dataset

OUT_FILE = "Spisok.jsonl"
TARGET_NON_TOXIC = 100                
TEXT_MIN_LEN = 20
TEXT_MAX_LEN = 200
DATASETS = [
    ("mteb/banking77", "train", "text", None, None),          
    ("toughdata/banking77", "train", "text", None, None),    
    ("amazon_polarity", "train", "text", "label", 1),         
    ("ag_news", "train", "text", None, None),                
    ("yahoo_answers_topics", "train", "text", None, None),
    ("imdb", "train", "text", "label", 1),                   
]

def clean_text(text):
    text = str(text).strip()
    text = text.replace("\n", " ")
    text = " ".join(text.split())
    return text

def is_good_text(text):
    return text and TEXT_MIN_LEN <= len(text) <= TEXT_MAX_LEN

all_texts = []

for repo, split, field, filter_field, filter_value in DATASETS:
    print(f"\nЗагрузка {repo} (split: {split})...")
    try:
        ds = load_dataset(repo, split=split, trust_remote_code=False)
    except Exception as e:
        print(f" Ошибка: {e}")
        continue
    if field not in ds.column_names:
        print(f"  Поле '{field}' отсутствует. Доступны: {ds.column_names}")
        continue
    if filter_field is not None and filter_field not in ds.column_names:
        print(f"  Поле фильтра '{filter_field}' отсутствует. Пропускаем фильтрацию.")
        filter_field = None

    count = 0
    for item in ds:
        if filter_field is not None:
            if item[filter_field] != filter_value:
                continue
        text = clean_text(item[field])
        if is_good_text(text):
            all_texts.append(text)
            count += 1
    print(f" Извлечено {count} текстов")

print(f"\nВсего собрано {len(all_texts)}.")

random.shuffle(all_texts)
seen = set()
if Path(OUT_FILE).exists():
    with open(OUT_FILE, "r", encoding="utf-8") as f:
        for line in f:
            try:
                rec = json.loads(line)
                if "text" in rec:
                    h = hashlib.md5(rec["text"].encode()).hexdigest()
                    seen.add(h)
            except:
                pass

result = []
for text in all_texts:
    h = hashlib.md5(text.encode()).hexdigest()
    if h in seen:
        continue
    seen.add(h)
    result.append({
        "text": text,
        "category": "non_toxic",
        "label": "non_toxic"
    })
    if len(result) >= TARGET_NON_TOXIC:
        break

with open(OUT_FILE, "a", encoding="utf-8") as f:
    for row in result:
            f.write(json.dumps(row, ensure_ascii=False) + "\n")
print(f"\n Добавлено {len(result)} нетоксичных примеров в {OUT_FILE}")

```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

file = "Spisok.jsonl"
data = []
with open(file, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            data.append(json.loads(line))

df = pd.DataFrame(data)

df["len"] = df["text"].str.len()
df["label"] = df["label"].str.lower()

print("=== Статистика ===")
print(df.groupby("label")["len"].agg(["count", "min", "max", "mean", "median"]))

df["label"].value_counts().plot(kind="bar")
plt.title("Баланс классов")
plt.show()

for lbl in df["label"].unique():
    df[df["label"] == lbl]["len"].hist(alpha=0.6, label=lbl)
plt.legend()
plt.title("Длина текста")
plt.show()

if "category" in df.columns:
    df[df["label"] == "toxic"]["category"].value_counts().plot(kind="bar")
    plt.title("Категории токсичности")
    plt.show()

print("\n=== Примеры ===")
for cat in df["category"].unique():
    print(f"\n{cat}:")
    for t in df[df["category"] == cat]["text"].head(3):
        print("  ", t[:120])
```
